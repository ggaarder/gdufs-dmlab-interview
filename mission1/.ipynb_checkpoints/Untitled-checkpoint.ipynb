```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def plotit(filename):
    plot = pd.read_csv(filename, sep="\t", header=None)
    plt.scatter(plot[0], plot[1], c=plot[2])
```

```python
plotit("task1.txt")
```

```python
plotit("task2.txt")
```

```python
plotit("task3.txt")
```
